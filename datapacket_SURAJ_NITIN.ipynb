###  RANDOM SAMPLING

```python
import random
def random_subset( iterator, K ):
    result = []
    N = 0

    for item in iterator:
        N += 1
        if len( result ) < K:
            result.append( item )
        else:
            s = int(random.random() * N)
            if s < K:
                result[ s ] = item

    return result
```

```python
# random function called to create 10 reservoir 
reservoir1=random_subset(flow_key_list,1000)
reservoir2=random_subset(flow_key_list,1000)
reservoir3=random_subset(flow_key_list,1000)
reservoir4=random_subset(flow_key_list,1000)
reservoir5=random_subset(flow_key_list,1000)
reservoir6=random_subset(flow_key_list,1000)
reservoir7=random_subset(flow_key_list,1000)
reservoir8=random_subset(flow_key_list,1000)
reservoir9=random_subset(flow_key_list,1000)
reservoir10=random_subset(flow_key_list,1000)

```

```python
### to find the 100 most frequent SRC_DST_IP_Pair
## using MG sketch method
```

```python
### maximum error in estimate = N/K where N is the stream size and K is the size of table
### for error to be 1%...taking k vaues as 100...which will also return the top most (100) frequent items
```

```python
### SRC_DST_IP......dataframe comnverted to list
a_list=df3.tolist()
len(a_list)
```

```python
## code for MG method and we are taking weight for minus as =1 if input is greater tha table size
MG={}
def misra_gries():
    k=100
    count = 0 
    decrements = 0
    for ip in a_list:
        if ip in MG:      #Case A - we increase an existing counter
            MG[ip] += 1
        elif count < k:   #Case B - we create a new counter
            MG[ip] = 1
            count += 1
        else:             #Case C - a decrement step - we decrease all entries by one and remove zeroed out counters. 
            decrements += 1
            for z,v in list(MG.items()): #list() convert to avoid annoying python types
                v -= 1
                if v == 0:
                    MG.pop(z)
                    count -= 1
                else:
                    MG[k] = v
       
```

```python
# calling MG function
misra_gries()
```

```python
### sorted frequent items
import operator
sorted_d = dict( sorted(MG.items(), key=operator.itemgetter(1),reverse=True))
sorted_d
```

```python
### final 100 answer with source and destination IP of most frequent items
term=0
for key,values in sorted_d.items():
    term+=1
    print(key,"---",values)
    if term == 100:
        break
```

```python
### combining SRC_DST_IP pair separated by  " dot(.)" eliminator
df_combine= df[['src_ip', 'dst_ip']].agg('.'.join, axis=1)
```

```python
#### so for probalistic approach we are takin intitution Of KMV method

###The intuition behind KMV is straightforward. Supposing you have a good hash function, i.e.
#hash values are evenly distributed over the hash space (I will normalize the hash space output to [0-1] for the rest of this),
#then you could estimate the number of distinct values you have seen by knowing the average spacing between values in 
#the hash space. If I see 10 distinct values, I would expect them on average to be spaced about 1/10th apart from each other.
#We could do this cheaply by keeping track of, say, the smallest value you have ever seen. 
#If the values are indeed uniformly distributed and provided youâ€™ve thrown a decent amount of data through it, 
#you could guess that the smallest value you have seen is a decent estimate of the average spacing of hash values in your space.
```

```python
# values of combine dataframe
df_combine
```

```python
### algorithm for KMV sketch method
### stored 1000 value intially as table size is less than k
k=1000
table=[]
for x in  range(0,1001):
    a=hash(df_combine[x])
    if a<0:               ### making negative hash value as positive
        b=(-1)*a
        table.append(b)
    else:
        b=a
        table.append(b)
```

```python
#last hatch value
table[999]
```

```python
table.pop(1000)  ### as intially we want table of size 1000
```

```python
for y in range(1000,2704839):
    l=hash(df_combine[y])         ### to compute hash value after table is created with intiall 1000 value
    if l<0:
           j=l*(-1)               ### negative hash value is computed as positive
    else:
           j=l
    table.sort
    if l<table[999]:           ### [length(table)-1]=999
        table.pop()
        table.append(l)
    
```

```python
value_1=table[999]*(-1)
table.pop(999)
table.append(value_1)
table.sort()

```

```python
### finally sort the table and do normalization to get a uniform value between 0 and 1 
### converting table list to pandas data frame
table_dataframe=pd.DataFrame(table,columns=["hatchvalue"])
```
